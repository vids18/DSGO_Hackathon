```python
df['date'] = df[['DATEON']].astype('datetime64')
df['year'] = df['date'].dt.year
df.sample(3)
```

```python
fig = px.scatter(df, x="year", y="TOTAL_NO3", hover_data=['TOTAL_NO3'])
fig.show()
```

```python
df.corrwith(df['TOTAL_NO3'])
```

```python
df3 = df.groupby(['SITE_ID','year']).agg({'TOTAL_NO3': lambda x: x.sum(skipna=True)})
```
